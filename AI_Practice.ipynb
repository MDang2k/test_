```python
def thresh(x):
    if x == 1:
        return 0
    elif x== 2: 
        return 1
    elif x== 3: 
        return 2
    elif x== 4: 
        return 3
    elif x== 5: 
        return 4
    elif x== 6: 
        return 5
    elif x== 7: 
        return 6
    elif x== 8: 
        return 7
    else: 
      return 8

thresh_v = np.vectorize(thresh)

def create_dataset(paths):
  
  mat_contents  = sio.loadmat(paths)
  imb = mat_contents['imdbOCTwithLabels'] 
  
  data = imb[0,0]['images']['data'][0,0]
  data = np.transpose(data, (3,2,0,1))
  
  label = imb[0,0]['images']['layerLabels_Exp1'][0,0]
  label = np.transpose(label, (3,2,0,1))
  label = thresh_v(label)
  

  return data, label
```

```python
x, y = create_dataset('C:/Users/Potato/Practice Data/imdbOCTwithLabels.mat')
print(x.shape,y.shape)
```
